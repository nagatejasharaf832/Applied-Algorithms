Here is the below code for this where I have used lambda, sum, and map to find sum of reciprocals where I have given a function 1/(int(a) where lambda function will automatically generates reciprocal of the given number and sum will add all the reciprocal numbers which gives us the result as sum of reciprocal of the given number.


```python
def f(*x,**y):
    def s1(x):
        s = 0
        if x:
            for i in x:
                s += i
        return s
        
    def p1(x):
        s = 1
        if x:
            for i in x:
                s*= i
        return s
    def recisum(x):
             return(sum(map(lambda a : (1/int(a)), x)))
    if y["action"] == "sum":
        return s1(*x)
    elif y["action"] == "prod":
        return p1(*x)
    elif y["action"] == "reciprocal sum":
        return recisum(*x)
    else:
        return f"bad argument: {y}"
if __name__ =='__main__':
    xlst = [1,2,3,4,5]
    print(f(xlst, action = "sum"))
    print(f(xlst, action = "prod"))
    print(f(xlst, action = "reciprocal sum"))
```

Task: Multiplication over complex number using operator overloading
Observation:
Here We are going to multiply two or more complex numbers using their formuloa
Here I have done this using operator overloading.

```python
class complex_:
    def __init__(self, re=0, im=0):
        self.re = re
        self.im = im
    def get_re(self):
        return self.re
    def get_im(self):
        return self.im
    def __str__(self):
        g = lambda x:"+" if x >= 0 else ""
        return f"({self.re}{g(self.im)}{self.im}i)"
    def cadd(self, other):
        new_re = self.get_re() * other.get_re()
        new_im = self.get_im() * other.get_im()
        return complex_(new_re, new_im)
    def __mul__(self, other):
#       complex numbers multiplication formula from http://www2.clarku.edu/faculty/djoyce/complex/mult.html
        return complex_((self.re*other.get_re()-self.im * other.get_im()),(self.re*other.get_im()+other.get_re()*self.im))
       
if __name__ =='__main__':
    w = complex_(1,-3)
    x = complex_(-1,3)
    y = complex_(1,3)
    z = complex_(-1,-3)
    print(w * x * y * z)
```

Observations:
Here given 3 functions we need to perform experimental analysis on run times and inputs given so that we need to figure out which algorithm is running faster.
Here I have using plot function to perform running time of these algorithms.
By the output I got I can say that 
alogorithm3>aglorithm2>algorithm3
So which means the running time of algorithms is given as:
average3>average2>average3


```python
import time
import numpy as np
import matplotlib.pyplot as plt
from numpy.lib.function_base import append

def average1(S):
    #S:sequence
    n = len(S)
    my_average = [0]*n
    for j in range(n):
        total = 0
        for i in range(j + 1):
            total += S[i]
        my_average[j] = total / (j+1)
    return my_average
def average2(S):
    #S:sequence
    n = len(S)
    my_average= [0]*n
    for j in range(n):
        my_average[j] = sum(S[0:j+1]) / (j+1)
    return my_average
def average3(S):
    #S:sequence
    n = len(S)
    my_average = [0]*n
    total = 0
    for j in range(n):
        total += S[j]
        my_average[j] = total / (j+1)
    return my_average


def run(fun,random_list):
    intial_time = time.perf_counter()
    fun(random_list)
    finishing_time = time.perf_counter()
    return (finishing_time-intial_time)

inp = []

times_1=[]
times_2=[]
times_3=[]
for i in range(10000,30000,50):
    inp.append(1*i)
res = []

for i in range(len(inp)):

    times_1.append(run(average1,inp[0:i]))
    times_2.append(run(average2,inp[0:i]))
    times_3.append(run(average3,inp[0:i]))
    res.append(len(inp[0:i]))

plt.yscale("log")
plt.xscale("log")
plt.plot(res,times_1,color = "r",label = "average1")
plt.plot(res,times_2,color = "b",label = "average2")
plt.plot(res,times_3,color = "g",label = "average3")
plt.xlabel("Inputs")
plt.ylabel("Run Time")
plt.title("log-log chart for various algorithms")
plt.legend(["average1","average2","average3"])
plt.show()



```
