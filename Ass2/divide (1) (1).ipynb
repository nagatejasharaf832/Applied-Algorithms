The approach's temporal complexity is O. (n).
The code's steps are listed below.
Take two array inputs.
2. 
From each array, remove and sort duplicate data (Remove duplicate because we need to check if both the array contains same set of unique elements or not)
3. 
Compare and contrast the sizes of each arrays.
4. 
If the sizes aren't the same, the arrays aren't the same.
5. 
If the size of the arrays is the same, we loop through any one of them and compare elements from both lists at the same index.
6. 
If the value is not the same at any index during the loop, the arrays are not the same, and the loop is terminated.
7. 
If the operation continues to the finish, both arrays will be the same.
O(n)

```python

def removeDuplicateNumbers(unsortedArr):
    unSortedArrlength = len(unsortedArr)
    length = int(max(max(unsortedArr),unSortedArrlength))
    B = []
    C = [0] * (length + 1)

    for j in range(0,unSortedArrlength):
        if C[unsortedArr[j]] == 0:
            C[unsortedArr[j]] = C[unsortedArr[j]] + 1

    for k in range(len(C)):
        if C[k] == 1:
            B.append(k)
    
    return B

def inputArrAndSort(D1, D2):
    D1_sorted = removeDuplicateNumbers(D1)
    D2_sorted = removeDuplicateNumbers(D2)
    
    areSame = False
    if len(D1_sorted) != len(D2_sorted):
        areSame = False
    else:
        for i in range(len(D1_sorted)):
            if D1_sorted[i] != D2_sorted[i]:
                areSame = False
                break
        areSame = True

    print("D1: ", D1)
    print("D2: ", D2)
    if areSame:
        print("The set D1 and D2 are the same")
    else:
        print("The set D1 and D2 are not the same")



# if __name__ == '__main__':
D1 = [3,1,2,1,10,12]
D2 = [2,3,1]

inputArrAndSort(D1,D2)

D1 = [3,1,2,1]
D2 = [2,3,1,2,3]

inputArrAndSort(D1,D2)
    
```

The Counting method with a time complexity of O can sort an array of size n with a range of [0, (n2) -1]. (n). The maximum number in the input list determines the size of the frequency array in the counting sort algorithm.
If we consider the list's size to be the same as the input array's size, there's a good chance we'll get an index out of bound error at runtime. As a result, we will utilize the size of the frequency array as (n2) to avoid the possibility of inaccuracy.
The counting algorithm has an O time complexity (n).
Please see the code below for further information.
O(n)

```python
import time
import random as rn
import math

def sortUsingCountingAlgo(unsortedArr, rangeOfInput):
    unSortedArrlength = len(unsortedArr)
    B = [0] * unSortedArrlength #Initializing the array with 0
    C = [0] * (rangeOfInput + 1) #Initializing the array with 0

    #Checking the occurrences of the number and incrementing the counter
    for j in range(0,unSortedArrlength):
        C[unsortedArr[j]] = C[unsortedArr[j]] + 1
    
    #Creating frequency depending on the occurrences of the number
    for i in range(1,len(C)):
        C[i] = C[i] + C[i - 1]

    #Sorting the array using the frequency array and the input array
    for k in range(unSortedArrlength - 1, -1, -1):
        B[C[unsortedArr[k]] - 1] = unsortedArr[k] 
        C[unsortedArr[k]] = C[unsortedArr[k]] - 1
        
    #Printing the sorted array
    print(B)

def calcRunTimeAndSort(arr, rangeOfInput):
    start = time.time()
    sortUsingCountingAlgo(arr, rangeOfInput)
    return time.time() - start  

# if _name_ == '_main_':
arrInput = []
n = 10
rangeOfInput = int(math.pow(n,2) - 1)
for i in range(n):
    arrInput.append(rn.randint(0,rangeOfInput))

print("Input Array")
print(arrInput)    
print("Array after sorting using counting algorithm")
print("Time taken to sort the array: ", calcRunTimeAndSort(arrInput, rangeOfInput))
```

The easiest technique to check if the array includes a duplicate entry in the given situation is to use hashmap.
We'll begin by traversing the array and inserting the value at the same place in the hashmap as the value.
We'll check if the element already exists at that index before entering the value.
We can return true if it exists; else, we'll continue with the rest of the array input.
We'll return false at the end of the for loop since if there were duplicate elements, we wouldn't have gotten this far in the program.
Dictionaries are used to implement hashmaps in Python.
O(n)

```python
import random as rn

def checkIfArrContainsDup(arrInput):
    dictionaryForInput = {}
    for i in range(len(arrInput)):
        temp = arrInput[i]
        if dictionaryForInput.get(temp) is None:
            dictionaryForInput[temp] = temp
        else:
            return True
    return False


# if _name_ == '_main_':
    
arrInput = []
for i in range(20):
    arrInput.append(rn.randint(0,100))

print("Input Array 1: ", arrInput)
print("Does input array contains duplicate elements: ", checkIfArrContainsDup(arrInput))

arrInput = [10,20,30,40,50,60]
print("Input Array 2: ", arrInput)
print("Does input array contains duplicate elements: ", checkIfArrContainsDup(arrInput))

arrInput = [10,20,20,40,50,50]
print("Input Array 3: ", arrInput)
print("Does input array contains duplicate elements: ", checkIfArrContainsDup(arrInput))
```

Heap sort is a one-of-a-kind strategy that ensures that all elements are sorted using a simple yet complex manner of sorting.
The heap-sort algorithm works by establishing a rigorous link between components and their arrangements. To begin, the original array is supplied to the build max heap method, which aids in pushing the highest element to the top of the array.
This is swapped with the last element and repeated until the max heapify function is invoked and the entire array has been sorted. The degree of difficulty remains O(n log n)
O(nlogn)

```python
import random
randomlist = random.sample(range(-20, 10), 10)
list_random = randomlist.copy()
def max_heapify(D,i,input_length):

    left = 2*i
    right = 2*i+1
    input_length = input_length-1
    if(left<=input_length and D[left]>D[i]):
        largest = left
    else:
        largest = i
    if(right<=input_length and D[right]>D[largest]):
        largest = right
    if(largest!=i):
        temp  = D[i]
        D[i] = D[largest]
        D[largest] = temp
        max_heapify(D,largest,input_length)
    return D

def build_max_heap(D):
    input_length = len(D)
    for i in range(input_length//2,1,-1):
        max_heapify(D,i-1,input_length)
    return D

def heap_sort(D):
    build_max_heap(D)
    input_length = len(D)
    input_length = input_length
    for i in range(input_length-1,1,-1):
        temp = D[1]
        D[1] = D[i]
        D[i] = temp
        input_length = input_length-1
        max_heapify(D,1,input_length)
    return D


D = [0,10, 16, 5, 8, 14, 9,0, 1, 18]
# D = [20, 16, 30, 15, 10, 8, 50]
x = heap_sort(list_random)
# x = max_heapify(D,0,len(D))
# x = build_max_heap(D)
print(x)

```

A counting sort is a sorting method that uses keys that fall inside a certain range. It operates by determining the number of items with unique key values (kind of hashing). The position of each object in the output sequence is then calculated using arithmetic.
O(n)

```python
import time

def sortUsingCountingAlgo(unsortedArr):
    unSortedArrlength = len(unsortedArr)
    B = [0] * unSortedArrlength
    C = [0] * (unSortedArrlength + 1)


    for j in range(0,unSortedArrlength):
        C[unsortedArr[j]] = C[unsortedArr[j]] + 1

    for i in range(1,len(C)):
        C[i] = C[i] + C[i - 1]

    for k in range(unSortedArrlength - 1, -1, -1):
        B[C[unsortedArr[k]] - 1] = unsortedArr[k] 
        C[unsortedArr[k]] = C[unsortedArr[k]] - 1
    
    print(B)

def calcRunTimeAndSort(arr):
    start = time.time()
    sortUsingCountingAlgo(arr)
    return time.time() - start  

# if __name__ == '__main__':
    
arrUnsorted = [5,4,2,1,4,1,10,4,5,7,5]
print("Array after sorting using counting algorithm")
print(calcRunTimeAndSort(arrUnsorted))
```
